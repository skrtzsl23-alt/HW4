```python
sites = [
"GAGGTAAAC",
"TCCGTAAGC",
"CAGGTTGGA",
"ACAGTCAGC",
"TAGGTCAGC",
"CAGGTCAGC",
"CAGGTCGAT",
"CAGGTCAGC",
"CAGGTCAGC",
"CAGGTTGGC"
]
```

```python
import numpy as np

al = ['A', 'T', 'G', 'C']
index = {n:i for i,n in enumerate(al)}

def create_pfm(sites):
    L = len(sites[0])
    pfm = np.zeros((4, L))

    for s in sites:
        for pos, nuc in enumerate(s):
            pfm[index[nuc], pos] += 1

    return pfm

pfm = create_pfm(sites)
print(pfm)
```

```python
def convert_to_ppm(pfm, alpha=0.1):
    N = pfm.sum(axis=0)[0] 
    ppm = (pfm + alpha) / (N + 4*alpha)
    return ppm

ppm = convert_to_ppm(pfm)
print(ppm)
```

```python
bg = np.array([0.295, 0.295, 0.205, 0.205]).reshape(4,1)

def calculate_pwm(ppm, bg):
    pwm = np.log2(ppm / bg)
    return pwm

pwm = calculate_pwm(ppm, bg)
print(pwm)
```

```python
def max_score(pwm):
    return pwm.max(axis=0).sum()

def min_score(pwm):
    return pwm.min(axis=0).sum()

print("max score =", max_score(pwm))
print("min score =", min_score(pwm))
```

```python
def best_seq(pwm, al):
    seq = ""
    for i in range(pwm.shape[1]):
        idx = np.argmax(pwm[:, i])
        seq += al[idx]
    return seq

print(best_seq(pwm, al))
```

```python
def worst_seq(pwm, al):
    seq = ""
    for i in range(pwm.shape[1]):
        idx = np.argmin(pwm[:, i])
        seq += al[idx]
    return seq

print(worst_seq(pwm, al))
```
