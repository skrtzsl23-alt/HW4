```python
# Transcription Factor Binding Sites Search in Genome
# Homework 4 - Task 2: PWM/PSSM scanning
# Author: skrtzsl23-alt

from Bio import motifs
from Bio.Seq import Seq
from Bio import SeqIO
import numpy as np
```

```python
# Input binding sites (known TF binding sequences)
binding_sites = [
    "GAGGTAAAC",
    "TCCGTAAGC",
    "CAGGTTGGA",
    "ACAGTCAGC",
    "TAGGTCAGC",
    "CAGGTCAGC",
    "CAGGTCGAT",
    "CAGGTCAGC",
    "CAGGTCAGC",
    "CAGGTTGGC"
]

# Convert to Seq objects as required by Bio.motifs
seq_objects = [Seq(seq) for seq in binding_sites]

# Create motif from sequences
motif = motifs.create(seq_objects)

# Extract Position Weight Matrix (PWM) and Position-Specific Scoring Matrix (PSSM)
position_weight_matrix = motif.pwm
pssm_matrix = motif.pssm
```

```python
# Display PWM for verification
print("=== Position Weight Matrix (PWM) ===")
for nucleotide, values in position_weight_matrix.items():
    print(f"{nucleotide}: {tuple(round(v, 4) for v in values)}")
```

```python
# Load genome from FASTA file (first 100,000 bases for analysis)
fasta_records = SeqIO.parse("chr1.fa", "fasta")
first_record = next(fasta_records)
full_genome = first_record.seq
genome_region = full_genome[:100000]  # Analyze initial segment

print(f"Genome region length: {len(genome_region)} bp")
```

```python
# Scan forward strand
forward_hits = []
score_threshold = 5.0

for position, score in pssm_matrix.search(genome_region, threshold=score_threshold):
    forward_hits.append((position, "+", score))

print(f"Found {len(forward_hits)} hits on forward strand (threshold >= {score_threshold})")
```

```python
# Scan reverse complement strand
reverse_genome = genome_region.reverse_complement()
reverse_hits = []

for position, score in pssm_matrix.search(reverse_genome, threshold=score_threshold):
    reverse_hits.append((position, "-", score))

print(f"Found {len(reverse_hits)} hits on reverse strand (threshold >= {score_threshold})")
```

```python
# Combine all hits
all_hits = forward_hits + reverse_hits
print(f"\n=== TOTAL HITS: {len(all_hits)} ===")
```

```python
# Display all hits with coordinates and scores
print("\n--- Detailed hit list (position, strand, score) ---")
for hit in all_hits:
    print(hit)
```

```python
# Additional statistics
if all_hits:
    scores = [hit[2] for hit in all_hits]
    print(f"\n=== SCORE STATISTICS ===")
    print(f"Maximum score: {max(scores):.4f}")
    print(f"Minimum score: {min(scores):.4f}")
    print(f"Average score: {np.mean(scores):.4f}")
    print(f"Median score: {np.median(scores):.4f}")
```
