```python
import numpy as np

# Position Probability Matrix from previous analysis
prob_matrix = np.array([
    [0.10576923, 0.77884615, 0.10576923, 0.00961538, 0.00961538, 0.20192308, 0.68269231, 0.20192308, 0.10576923],
    [0.20192308, 0.00961538, 0.00961538, 0.00961538, 0.97115385, 0.20192308, 0.00961538, 0.00961538, 0.10576923],
    [0.10576923, 0.00961538, 0.77884615, 0.97115385, 0.00961538, 0.00961538, 0.29807692, 0.77884615, 0.00961538],
    [0.58653846, 0.20192308, 0.10576923, 0.00961538, 0.00961538, 0.58653846, 0.00961538, 0.00961538, 0.77884615]
])


def compute_shannon_entropy(prob_matrix):
    """Calculate Shannon entropy for each position of the motif."""
    entropy_values = []
    for col_idx in range(prob_matrix.shape[1]):
        col_entropy = 0
        for prob in prob_matrix[:, col_idx]:
            if prob > 0:
                col_entropy -= prob * np.log2(prob)
        entropy_values.append(col_entropy)
    return np.array(entropy_values)

entropy = compute_shannon_entropy(prob_matrix)
entropy
```

```python
# Information content (maximum entropy 2 bits minus observed entropy)
max_entropy = 2
info_content = max_entropy - entropy
info_content
```

```python
import matplotlib.pyplot as plt

motif_positions = np.arange(1, prob_matrix.shape[1] + 1)

# Create bar plot of information content
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(motif_positions, info_content, color='steelblue', edgecolor='navy', alpha=0.8)

# Customize the plot
ax.set_xlabel('Motif Position', fontsize=12)
ax.set_ylabel('Information Content (bits)', fontsize=12)
ax.set_title('Sequence Logo Information Content by Position', fontsize=14, fontweight='bold')
ax.set_xticks(motif_positions)
ax.set_ylim(0, 2.1)
ax.axhline(y=2, color='red', linestyle='--', alpha=0.5, label='Maximum possible (2 bits)')
ax.legend()

# Add value labels on top of bars
for bar, height in zip(bars, info_content):
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.05,
            f'{height:.2f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

**Key observations:**
- **Core motif**: Positions with high IC ($\approx$2) indicate strong nucleotide preference → critical for binding
- **Variable positions**: Positions with low IC ($\approx$0–1) show more even probability distribution → less conserved
- Position 4 has lowest IC (0.23), suggesting this position is highly degenerate
