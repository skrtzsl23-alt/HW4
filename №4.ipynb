```python
import numpy as np
import matplotlib.pyplot as plt

# Position Probability Matrix from previous analysis
prob_matrix = np.array([
    [0.10576923, 0.77884615, 0.10576923, 0.00961538, 0.00961538, 0.20192308, 0.68269231, 0.20192308, 0.10576923],
    [0.20192308, 0.00961538, 0.00961538, 0.00961538, 0.97115385, 0.20192308, 0.00961538, 0.00961538, 0.10576923],
    [0.10576923, 0.00961538, 0.77884615, 0.97115385, 0.00961538, 0.00961538, 0.29807692, 0.77884615, 0.00961538],
    [0.58653846, 0.20192308, 0.10576923, 0.00961538, 0.00961538, 0.58653846, 0.00961538, 0.00961538, 0.77884615]
])


def compute_entropy(prob_matrix):
    """Calculate Shannon entropy for each position."""
    entropy_vals = []
    for col in range(prob_matrix.shape[1]):
        h = 0
        for p in prob_matrix[:, col]:
            if p > 0:
                h -= p * np.log2(p)
        entropy_vals.append(h)
    return np.array(entropy_vals)

entropy = compute_entropy(prob_matrix)
entropy
```

```python
# Information content (max entropy 2 bits minus observed entropy)
max_entropy = 2
info_content = max_entropy - entropy
info_content
```

```python
# Plot information content
positions = np.arange(1, prob_matrix.shape[1] + 1)

plt.figure(figsize=(10, 5))
plt.bar(positions, info_content, color='cornflowerblue', edgecolor='navy')
plt.xlabel('Motif position')
plt.ylabel('Information content (bits)')
plt.title('Information content per position')
plt.xticks(positions)
plt.ylim(0, 2.1)
plt.axhline(y=2, color='red', linestyle='--', alpha=0.7, label='Maximum (2 bits)')
plt.legend()
plt.show()
```

**Interpretation:**
- High IC (~2 bits): conserved positions critical for binding
- Low IC (0-1 bits): variable positions that tolerate substitutions
- Position 4 has the lowest IC (0.23), indicating high variability
