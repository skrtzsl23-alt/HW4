```python
import numpy as np
import matplotlib.pyplot as plt

PPM = np.array([
    [0.10576923, 0.77884615, 0.10576923, 0.00961538, 0.00961538, 0.20192308, 0.68269231, 0.20192308, 0.10576923],
    [0.20192308, 0.00961538, 0.00961538, 0.00961538, 0.97115385, 0.20192308, 0.00961538, 0.00961538, 0.10576923],
    [0.10576923, 0.00961538, 0.77884615, 0.97115385, 0.00961538, 0.00961538, 0.29807692, 0.77884615, 0.00961538],
    [0.58653846, 0.20192308, 0.10576923, 0.00961538, 0.00961538, 0.58653846, 0.00961538, 0.00961538, 0.77884615]
])

def shannon_entropy(ppm):
    H = []
    for i in range(ppm.shape[1]):
        h = 0
        for p in ppm[:, i]:
            if p > 0:
                h -= p * np.log2(p)
        H.append(h)
    return np.array(H)

H = shannon_entropy(PPM)
H
```

```python
IC = 2 - H
IC
```

```python
import matplotlib.pyplot as plt

positions = np.arange(1, PPM.shape[1] + 1)

plt.figure(figsize=(10, 5))
plt.bar(positions, IC, color='skyblue', edgecolor='navy')
plt.xlabel('Motif position')
plt.ylabel('Information content (bits)')
plt.title('Information content per position')
plt.xticks(positions)
plt.ylim(0, 2.1)
plt.axhline(y=2, color='red', linestyle='--', alpha=0.7, label='Maximum (2 bits)')
plt.legend()
plt.show()
```

Kernel matugu: positions with high IC (~2 bits) -> important for binding
Variable positions: low IC (0-1 bits) -> less conserved
