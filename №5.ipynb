```python
from Bio import SeqIO
import numpy as np
import matplotlib.pyplot as plt

records = SeqIO.parse("chr1.fa", "fasta")
record = next(records)
genome = record.seq
seq = genome[:1_000_000]
```

```python
nucleotides = ['A', 'T', 'G', 'C']
index = {nuc: i for i, nuc in enumerate(nucleotides)}
```

```python
dinuc_counts = np.zeros((4, 4), dtype=int)
for i in range(len(seq) - 1):
    a, b = seq[i], seq[i+1]
    if a in nucleotides and b in nucleotides:
        dinuc_counts[index[a], index[b]] += 1
```

```python
print("Частоты динуклеотидов (матрица счетов):")
print(dinuc_counts)
```

```python
transition_matrix = dinuc_counts / dinuc_counts.sum(axis=1, keepdims=True)
```

```python
print("\nМатрица переходов (Pij):")
print(np.round(transition_matrix, 4))
```

```python
transition_matrix
```

```python
row_sums = transition_matrix.sum(axis=1)
print("\nСумма строк матрицы переходов (должна быть 1):")
print(np.round(row_sums, 4))
```

```python
eigvals, eigvecs = np.linalg.eig(transition_matrix.T)
stationary = np.real(eigvecs[:, np.isclose(eigvals, 1)])
stationary = stationary / stationary.sum()  
stationary = stationary.flatten()
```

```python
print("\nСтационарное распределение")
for nuc, freq in zip(nucleotides, stationary):
    print(f"{nuc}: {freq:.4f}")
```

```python
obs_counts = np.array([seq.count(nuc) for nuc in nucleotides])
obs_freq = obs_counts / obs_counts.sum()
print("\nНаблюдаемые частоты нуклеотидов:")
for nuc, freq in zip(nucleotides, obs_freq):
    print(f"{nuc}: {freq:.4f}")
```

```python
print("\nСравнение стационарного распределения с наблюдаемыми частотами:")
for nuc, s, o in zip(nucleotides, stationary, obs_freq):
    print(f"{nuc}: π={s:.4f}, наблюдаемая={o:.4f}")
```
