```python
P = transition_matrix
```

```python
P_prev = P.copy()
P_k = P.copy()  
norm_diff = []
for k in range(2, 1001):
    P_k = P_k @ P  
    diff = np.linalg.norm(P_k - P_prev, ord='fro')
    norm_diff.append(diff)
    P_prev = P_k.copy()
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(2, 1001), norm_diff)
plt.xlabel('k (степень матрицы)')
plt.ylabel('||P^k - P^(k-1)||_F')
plt.title('Сходимость P^k к стационарной матрице')
plt.grid(True)
plt.show()
```
