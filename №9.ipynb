```python
from Bio import SeqIO
import numpy as np
from scipy.stats import chi2

records = SeqIO.parse("chr1.fa", "fasta")
record = next(records)
genome = record.seq
seq = str(genome[:100000]) 

alphabet = ["A", "C", "G", "T"]
index = {nt:i for i, nt in enumerate(alphabet)}
```

```python
def observed_dinucleotides(seq):
    O = np.zeros((4,4))
    for i in range(len(seq)-1):
        if seq[i] not in index or seq[i+1] not in index:
            continue 
        x = index[seq[i]]
        y = index[seq[i+1]]
        O[x, y] += 1
    return O

O_real = observed_dinucleotides(seq)
O_real
```

```python
counts = np.array([seq.count(nt) for nt in alphabet])
P = counts / sum(counts)
N = len(seq) - 1  


E = np.outer(P, P) * N
E 
```

```python
chi2_stat = np.sum((O_real - E)**2 / E)
df = 9  
p_value = 1 - chi2.cdf(chi2_stat, df)

print("χ² =", chi2_stat)
print("p-value =", p_value)
```

```python
np.random.seed(42)  
seq_random = ''.join(np.random.choice(alphabet, size=len(seq), p=P))

O_random = observed_dinucleotides(seq_random)
E_random = np.outer(P, P) * (len(seq_random) - 1)

chi2_random = np.sum((O_random - E_random)**2 / E_random)
p_value_random = 1 - chi2.cdf(chi2_random, df)

print("Случайная последовательность 0-го порядка:")
print("χ² =", chi2_random)
print("p-value =", p_value_random)
```
